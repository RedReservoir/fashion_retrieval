```python
import torch
import torchvision

import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

# Model specs

```python
model_data_df = pd.DataFrame({
    "model": pd.Series(dtype=str),
    "version": pd.Series(dtype=str),
    "gflops": pd.Series(dtype=float),
    "params": pd.Series(dtype=int),
    "acc@1": pd.Series(dtype=float)
})
```

```python
# Taken from PyTorch

model_name = "ResNet"
model_versions = ["18", "34", "50", "101", "152"]
model_weights = [
    torchvision.models.ResNet18_Weights.DEFAULT,
    torchvision.models.ResNet34_Weights.DEFAULT,
    torchvision.models.ResNet50_Weights.DEFAULT,
    torchvision.models.ResNet101_Weights.DEFAULT,
    torchvision.models.ResNet152_Weights.DEFAULT
]

#

for model_version, model_weights in zip(model_versions, model_weights):

    row = {
        "model": model_name,
        "version": model_version,
        "params": model_weights.meta["num_params"],
        "acc@1": model_weights.meta["_metrics"]["ImageNet-1K"]["acc@1"]
    }

    model_data_df.loc[len(model_data_df)] = row
```

```python
# Taken from HuggingFace
# Metrics from Table 2 in https://arxiv.org/pdf/2102.12122

# Zetatech/pvt-tiny-224
# Zetatech/pvt-small-224
# Zetatech/pvt-medium-224
# Zetatech/pvt-large-224

#

model_name = "PVT"
model_versions = ["T", "S", "M", "L"]
model_num_params_list = [13.2, 24.5, 44.2, 61.4]
model_acc_1_list = [75.1, 79.8, 81.2, 81.7]

#

for model_version, model_num_params, model_acc_1 in zip(model_versions, model_num_params_list, model_acc_1_list):

    row = {
        "model": model_name,
        "version": model_version,
        "params": model_num_params * 1000000,
        "acc@1": model_acc_1
    }

    model_data_df.loc[len(model_data_df)] = row
```

```python
# Taken from HuggingFace
# Metrics from Table 5 in https://arxiv.org/abs/2012.12877

# facebook/deit-tiny-patch16-224
# facebook/deit-tiny-distilled-patch16-224
# facebook/deit-small-patch16-224
# facebook/deit-small-distilled-patch16-224
# facebook/deit-base-patch16-224
# facebook/deit-base-distilled-patch16-224
# facebook/deit-base-patch16-384
# facebook/deit-base-distilled-patch16-384

# DeiT-Ti
# DeiT-Ti$
# DeiT-S
# DeiT-S$
# DeiT-B
# DeiT-B$
# Deit-B
# DeiT-B$|384/1000

#

model_name = "DeiT"
model_versions = ["T", "T-d", "S", "S-d", "B", "B-d", "B-384", "B-d-384"]
model_num_params_list = [5, 6, 22, 22, 86, 87, 87, 88]
model_acc_1_list = [72.2, 74.5, 79.9, 81.2, 81.8, 83.4, 82.9, 85.2]

#

for model_version, model_num_params, model_acc_1 in zip(model_versions, model_num_params_list, model_acc_1_list):

    row = {
        "model": model_name,
        "version": model_version,
        "params": model_num_params * 1000000,
        "acc@1": model_acc_1
    }

    model_data_df.loc[len(model_data_df)] = row
```

```python
# Taken from HuggingFace
# Metrics from Table 3 in https://arxiv.org/pdf/2103.15808

# microsoft/cvt-13
# microsoft/cvt-13-384
# microsoft/cvt-13-384-22k
# microsoft/cvt-21
# microsoft/cvt-21-384
# microsoft/cvt-21-384-22k
# microsoft/cvt-w24-384-22k

# CvT-13
# CvT-13|384
# CvT-13|384 (22k)
# CvT-21
# CvT-21|384
# CvT-21|384 (22k)
# CvT-W24|384 (22k)

#

model_name = "CvT"
model_versions = ["13", "13-384", "13-384-22k", "21", "21-384", "21-384-22k", "w24-384-22k"]
model_num_params_list = [20, 20, 20, 32, 32, 32, 277]
model_acc_1_list = [81.6, 83.0, 83.3, 82.5, 83.3, 84.9, 87.7]

#

for model_version, model_num_params, model_acc_1 in zip(model_versions, model_num_params_list, model_acc_1_list):

    row = {
        "model": model_name,
        "version": model_version,
        "params": model_num_params * 1000000,
        "acc@1": model_acc_1
    }

    model_data_df.loc[len(model_data_df)] = row
```

```python
model_names = [
    "ResNet",
    "EfficientNet",
    "ConvNeXt",
    "ViT",
    "SWinT",
    "PVT",
    "DeiT",
    "CvT"
]

#

plt.figure(figsize=(25, 5))

#

for model_idx, model_name in enumerate(model_names):

    model_data_filter_df = model_data_df[model_data_df["model"] == model_name]

    plt.axhline(-model_idx, xmin=0.01, xmax=0.99, color='black', linestyle='--', linewidth=0.5, zorder=0)

    plt.scatter(model_data_filter_df["params"], [-model_idx] * len(model_data_filter_df), label=model_name)

    for version, params in zip(model_data_filter_df["version"], model_data_filter_df["params"]):
        label = version
        x_pos = params
        y_pos = -model_idx + 0.15
        if model_name == "ViT":
            if "32" in version:
                y_pos += 0.40
        if model_name == "SWinT":
            if "V2" in version:
                y_pos += 0.40
        if model_name == "DeiT":
            if "B" in version:
                x_pos /= 1.075
            if "(d)" in version:
                y_pos += 0.40
            if "384" in version:
                x_pos *= 1.15
        plt.text(x_pos, y_pos, s=label, fontsize=12, ha="center")

#

plt.legend(loc=1)

plt.title("Num params")

plt.gca().get_yaxis().set_visible(False)
plt.xscale("log")

plt.ylim(-len(model_names) - 0.50 + 1, 0.75)

plt.show()
```
