```python
# Manual data

data_df = pd.DataFrame({
    "exp_name": pd.Series(dtype="str"),
    "neg_spl": pd.Series(dtype="str"),
    "rd_mode": pd.Series(dtype="str"),
    "autocast": pd.Series(dtype="str"),
    "ddp_order": pd.Series(dtype="str"),
    "base_lr": pd.Series(dtype="str"),
    "val_loss": pd.Series(dtype="float")
})

neg_spl_list = [
    "new",
    "old",
    "old",
    "old",
    "new",
    "new",
    "new",
    "new",
    "old",
    "old"
]

rd_mode_list = [
    "fixed",
    "fixed",
    "random",
    "none",
    "fixed",
    "fixed",
    "fixed",
    "fixed",
    "fixed",
    "fixed"
]

autocast_list = [
    "no",
    "no",
    "no",
    "no",
    "yes",
    "no",
    "no",
    "yes",
    "no",
    "yes"
]

ddp_order_list = [
    "new",
    "new",
    "new",
    "new",
    "new",
    "old",
    "new",
    "new",
    "new",
    "new"
]

base_lr_list = [
    1e-2,
    1e-2,
    1e-2,
    1e-2,
    1e-2,
    1e-2,
    6.25e-4,
    6.25e-4,
    6.25e-4,
    6.25e-4
]

fig, axs = plt.subplots(nrows=2, figsize=(20, 15))

# Baseline (REPRODUCTION)

exp_data = exp_data_all["ret_train"]["convnext_t"]["test_002"]

train_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["train_mean_loss_list"]]
val_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["val_mean_loss_list"]]

t_idx = min(len(train_loss_list), 5)
v_idx = min(len(val_loss_list), 5)

train_loss_list = train_loss_list[:t_idx]
val_loss_list = val_loss_list[:v_idx]

label = "Baseline (REPRODUCTION)"

axs[0].plot(range(1, len(train_loss_list) + 1), train_loss_list, label=label, color="gray", marker=".", linestyle="--")
axs[1].plot(range(1, len(val_loss_list) + 1), val_loss_list, label=label, color="gray", marker=".", linestyle="--")

row = {
    "exp_name": "Baseline (REPRODUCTION)",
    "val_loss": val_loss_list[-1]
}

data_df.loc[len(data_df)] = row

# Baseline (ORIGINAL)

exp_data = exp_data_all["ret_train"]["convnext_t"]["test_002_copy"]

train_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["train_mean_loss_list"]][:5]
val_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["val_mean_loss_list"]][:5]

label = "Baseline (ORIGINAL)"

axs[0].plot(range(1, len(train_loss_list) + 1), train_loss_list, label=label, color="black", marker=".", linestyle="--")
axs[1].plot(range(1, len(val_loss_list) + 1), val_loss_list, label=label, color="black", marker=".", linestyle="--")

row = {
    "exp_name": "Baseline (ORIGINAL)",
    "val_loss": val_loss_list[-1]
}

data_df.loc[len(data_df)] = row

# Experiments

for idx, (exp_name, exp_data) in enumerate(exp_data_all["change_test"]["convnext_t_s1"].items()):

    train_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["train_mean_loss_list"]]
    val_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["val_mean_loss_list"]]

    finished = exp_data["results"]["stage_1"]["finished"]

    # Title

    title_lines = []
    title_lines.append("{:s}".format(exp_name))
    if not finished: title_lines[-1] += " | Unifinished"
    title_lines.append("{:10s}: {:s}".format("NEG SPL", neg_spl_list[idx]))
    title_lines.append("{:10s}: {:s}".format("RND SHFL", rd_mode_list[idx]))
    title_lines.append("{:10s}: {:s}".format("AUTOCAST", autocast_list[idx]))
    title_lines.append("{:10s}: {:s}".format("DDP ORD", ddp_order_list[idx]))
    title_lines.append("{:10s}: {:s}".format("BASE LR", np.format_float_scientific(base_lr_list[idx], min_digits=2, exp_digits=1)))
    
    label = "\n".join(title_lines)

    # Color

    color_type = "basic"
    #color_type = "neg_smp"
    #color_type = "rd_mode"
    #color_type = "autocast"
    #color_type = "ddp_order"
    #color_type = "base_lr"

    if color_type == "basic":

        ## Basic
        color = tab20_color_list[idx]

    if color_type == "neg_smp":

        ## neg_smp
        color_dict = {
            "new": 0,
            "old": 1
        } 
        color = color_dict[neg_spl_list[idx]]

    if color_type == "rd_mode":

        ## rd_mode
        color_dict = {
            "fixed": 0,
            "random": 1,
            "none": 2
        } 
        color = color_dict[rd_mode_list[idx]]

    if color_type == "autocast":

        ## autocast
        color_dict = {
            "no": 0,
            "yes": 1
        } 
        color = color_dict[autocast_list[idx]]

    if color_type == "ddp_order":

        ## ddp_order
        color_dict = {
            "new": 0,
            "old": 1
        } 
        color = color_dict[ddp_order_list[idx]]

    if color_type == "base_lr":

        ## base_lr
        color_dict = {
            1e-2: 0,
            6.25e-4: 1
        } 
        color = color_dict[base_lr_list[idx]]

    # Plot

    axs[0].plot(range(1, len(train_loss_list) + 1), train_loss_list, label=label, marker=".", color=color)
    axs[1].plot(range(1, len(val_loss_list) + 1), val_loss_list, label=label, marker=".", color=color)
    
    # df

    row = {
        "exp_name": exp_name,
        "neg_spl": neg_spl_list[idx],
        "rd_mode": rd_mode_list[idx],
        "autocast": autocast_list[idx],
        "ddp_order": ddp_order_list[idx],
        "base_lr": np.format_float_scientific(base_lr_list[idx], min_digits=2, exp_digits=1),
        "val_loss": val_loss_list[-1]
    }

    data_df.loc[len(data_df)] = row

# Generic settings

axs[0].set_yscale("log")
axs[1].set_yscale("log")

axs[0].grid()
axs[1].grid()

axs[0].set_title("Train loss")
axs[1].set_title("Val loss")

plt.suptitle("Reproduction of past ConvNeXt-T runs")
plt.legend(prop={'family': 'monospace', "size": 13}, bbox_to_anchor=(1.20, 2.20))
plt.show()
```

```python
data_df = pd.DataFrame({
    "exp_name": pd.Series(dtype="str"),
    "order": pd.Series(dtype="str"),
    "speed": pd.Series(dtype="str"),
    "mode": pd.Series(dtype="str"),
    "weight": pd.Series(dtype="str"),
    "val_loss_5": pd.Series(dtype="float"),
    "val_loss_last": pd.Series(dtype="float"),
    "num_params": pd.Series(dtype="int"),
    "num_epochs": pd.Series(dtype="int")
})

for idx, (exp_name, exp_data) in enumerate(exp_data_all["ret_train_pyr_2"]["convnext_t_s1"].items()):

    val_loss_5 = sum(exp_data["results"]["stage_1"]["val_mean_loss_list"][4])
    val_loss_last = sum(exp_data["results"]["stage_1"]["val_mean_loss_list"][-1])
    num_epochs = exp_data["results"]["stage_1"]["num_epochs"]

    backbone_comp = src.utils.comps.create_backbone(exp_data["settings"]["backbone"])
    head_comp = src.utils.comps.create_head(backbone_comp, exp_data["settings"]["head"])
    num_params = src.utils.module.get_num_params(head_comp)

    head_class = exp_data["settings"]["head"]["class"]

    head_order = "BU" if "BottomUp" in head_class else "TD"
    head_speed = "Inst" if "Instant" in head_class else "Prog"
    head_du_mode = "Simple" if "Simple" in head_class else "Conv"
    if head_speed == "Prog":
        head_weight = "Early" if exp_data["settings"]["head"]["emb_sizes"][0] == 512 else "Late"
    else:
        head_weight = None
        
    row = {
        "exp_name": "exp_{:03d}".format(idx),
        "order": head_order,
        "speed": head_speed,
        "mode": head_du_mode,
        "weight": head_weight,
        "val_loss_5": val_loss_5,
        "val_loss_last": val_loss_last,
        "num_params": num_params,
        "num_epochs": num_epochs
    }

    data_df.loc[len(data_df)] = row
```
