```python
import torch
import torchvision

import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

# Model specs

```python
model_data_df = pd.DataFrame({
    "model": pd.Series(dtype=str),
    "version": pd.Series(dtype=str),
    "gflops": pd.Series(dtype=float),
    "params": pd.Series(dtype=int),
    "acc@1": pd.Series(dtype=float)
})
```

```python
# Taken from PyTorch

model_name = "ResNet"
model_versions = ["18", "34", "50", "101", "152"]
model_weights = [
    torchvision.models.ResNet18_Weights.DEFAULT,
    torchvision.models.ResNet34_Weights.DEFAULT,
    torchvision.models.ResNet50_Weights.DEFAULT,
    torchvision.models.ResNet101_Weights.DEFAULT,
    torchvision.models.ResNet152_Weights.DEFAULT
]

#

for model_version, model_weights in zip(model_versions, model_weights):

    row = {
        "model": model_name,
        "version": model_version,
        "params": model_weights.meta["num_params"],
        "acc@1": model_weights.meta["_metrics"]["ImageNet-1K"]["acc@1"]
    }

    model_data_df.loc[len(model_data_df)] = row
```

```python
# Taken from HuggingFace
# Metrics from Table 2 in https://arxiv.org/pdf/2102.12122

# Zetatech/pvt-tiny-224
# Zetatech/pvt-small-224
# Zetatech/pvt-medium-224
# Zetatech/pvt-large-224

#

model_name = "PVT"
model_versions = ["T", "S", "M", "L"]
model_num_params_list = [13.2, 24.5, 44.2, 61.4]
model_acc_1_list = [75.1, 79.8, 81.2, 81.7]

#

for model_version, model_num_params, model_acc_1 in zip(model_versions, model_num_params_list, model_acc_1_list):

    row = {
        "model": model_name,
        "version": model_version,
        "params": model_num_params * 1000000,
        "acc@1": model_acc_1
    }

    model_data_df.loc[len(model_data_df)] = row
```

```python
# Taken from HuggingFace
# Metrics from Table 5 in https://arxiv.org/abs/2012.12877

# facebook/deit-tiny-patch16-224
# facebook/deit-tiny-distilled-patch16-224
# facebook/deit-small-patch16-224
# facebook/deit-small-distilled-patch16-224
# facebook/deit-base-patch16-224
# facebook/deit-base-distilled-patch16-224
# facebook/deit-base-patch16-384
# facebook/deit-base-distilled-patch16-384

# DeiT-Ti
# DeiT-Ti$
# DeiT-S
# DeiT-S$
# DeiT-B
# DeiT-B$
# Deit-B
# DeiT-B$|384/1000

#

model_name = "DeiT"
model_versions = ["T", "T-d", "S", "S-d", "B", "B-d", "B-384", "B-d-384"]
model_num_params_list = [5, 6, 22, 22, 86, 87, 87, 88]
model_acc_1_list = [72.2, 74.5, 79.9, 81.2, 81.8, 83.4, 82.9, 85.2]

#

for model_version, model_num_params, model_acc_1 in zip(model_versions, model_num_params_list, model_acc_1_list):

    row = {
        "model": model_name,
        "version": model_version,
        "params": model_num_params * 1000000,
        "acc@1": model_acc_1
    }

    model_data_df.loc[len(model_data_df)] = row
```

```python
# Taken from HuggingFace
# Metrics from Table 3 in https://arxiv.org/pdf/2103.15808

# microsoft/cvt-13
# microsoft/cvt-13-384
# microsoft/cvt-13-384-22k
# microsoft/cvt-21
# microsoft/cvt-21-384
# microsoft/cvt-21-384-22k
# microsoft/cvt-w24-384-22k

# CvT-13
# CvT-13|384
# CvT-13|384 (22k)
# CvT-21
# CvT-21|384
# CvT-21|384 (22k)
# CvT-W24|384 (22k)

#

model_name = "CvT"
model_versions = ["13", "13-384", "13-384-22k", "21", "21-384", "21-384-22k", "w24-384-22k"]
model_num_params_list = [20, 20, 20, 32, 32, 32, 277]
model_acc_1_list = [81.6, 83.0, 83.3, 82.5, 83.3, 84.9, 87.7]

#

for model_version, model_num_params, model_acc_1 in zip(model_versions, model_num_params_list, model_acc_1_list):

    row = {
        "model": model_name,
        "version": model_version,
        "params": model_num_params * 1000000,
        "acc@1": model_acc_1
    }

    model_data_df.loc[len(model_data_df)] = row
```

```python
model_names = [
    "ResNet",
    "EfficientNet",
    "ConvNeXt",
    #"ViT",
    #"SWinT",
    #"PVT",
    #"DeiT",
    #"CvT"
]

model_colors = {
    "ResNet": plt.get_cmap("tab10")(0.05),
    "EfficientNet": plt.get_cmap("tab10")(0.15),
    "ConvNeXt": plt.get_cmap("tab10")(0.25),
    #"ViT": plt.get_cmap("tab10")(0.35),
    #"SWinT": plt.get_cmap("tab10")(0.45),
    #"PVT": plt.get_cmap("tab10")(0.55),
    #"DeiT": plt.get_cmap("tab10")(0.65),
    #"CvT": plt.get_cmap("tab10")(0.75)
}

model_alphas = {
    "ResNet": 1,
    "EfficientNet": 1,
    "ConvNeXt": 1,
    #"ViT": 1,
    #"SWinT": 1,
    #"PVT": 1,
    #"DeiT": 1,
    #"CvT": 1
}

#

model_data_ord_df = model_data_df.copy()

#

plt.figure(figsize=(20, 10))

for model_idx, model_name in enumerate(model_names):

    color = model_colors[model_name]
    alpha = model_alphas[model_name]

    model_data_filter_df = model_data_ord_df[model_data_ord_df["model"] == model_name]

    plt.scatter(model_data_filter_df["params"], model_data_filter_df["acc@1"], label=model_name, alpha=alpha, color=color)

    for version, params, acc_1 in zip(model_data_filter_df["version"], model_data_filter_df["params"], model_data_filter_df["acc@1"]):
        label = version
        x_pos = params
        y_pos = acc_1 + 0.3
        if model_name == "SWinT" and "V2" not in version:
            y_pos = acc_1 - 0.65
        plt.text(x_pos, y_pos, s=label, fontsize=10, ha="center", alpha=alpha)
        
#

plt.xscale("log")

lgnd = plt.legend(loc="lower right")
for handle in lgnd.legend_handles:
    handle._sizes = [30]

plt.xlabel("# Params")
plt.ylabel("ImageNet-1K acc@1")

plt.xlim(model_data_df["params"].min() / 10**(0.1), model_data_df["params"].max() * 10**(0.1))
plt.ylim(model_data_df["acc@1"].min() - 1, model_data_df["acc@1"].max() + 1.5)

#

model_and_version_list = [
    ("EfficientNet", "S-V2"),
    ("EfficientNet", "M-V2"),
    ("EfficientNet", "L-V2")
]

x_lth = 0.065
y_lth = 1.1

for model, version in model_and_version_list:

    row = model_data_df[(model_data_df["model"] == model) & (model_data_df["version"] == version)]
    params = row.at[row.index[0], "params"]
    acc_1 = row.at[row.index[0], "acc@1"]

    x_cen = np.log10(params)
    y_cen = acc_1

    x_pos = x_cen - (x_lth / 2)
    y_pos = y_cen - (y_lth / 2) + 0.20

    x_pos_exp = 10**x_pos
    x_lth_exp = 10**(x_pos+x_lth) - 10**x_pos

    color = model_colors[model]
    rect = mpl.patches.Rectangle((x_pos_exp, y_pos), x_lth_exp, y_lth, linewidth=2, edgecolor=color, facecolor='none')
    plt.gca().add_patch(rect)

#

plt.grid()
plt.show()
```

```python
model_names = [
    "ResNet",
    "EfficientNet",
    "ConvNeXt",
    "ViT",
    "DeiT",
    "PVT",
    "CvT",
    "SWinT"
]

model_colors = {
    "ResNet": plt.get_cmap("tab10")(0.05),
    "EfficientNet": plt.get_cmap("tab10")(0.15),
    "ConvNeXt": plt.get_cmap("tab10")(0.25),
    "ViT": plt.get_cmap("tab10")(0.35),
    "DeiT": plt.get_cmap("tab10")(0.45),
    "PVT": plt.get_cmap("tab10")(0.55),
    "CvT": plt.get_cmap("tab10")(0.65),
    "SWinT": plt.get_cmap("tab10")(0.75)
}

model_alphas = {
    "ResNet": 0.33,
    "EfficientNet": 0.33,
    "ConvNeXt": 0.33,
    "ViT": 1,
    "DeiT": 1,
    "PVT": 1,
    "CvT": 1,
    "SWinT": 1
}

#

model_data_ord_df = model_data_df.copy()

model_and_version_list = [
    ("ResNet", "18"),
    ("ResNet", "34"),
    ("EfficientNet", "B0"),
    ("EfficientNet", "B1"),
    ("EfficientNet", "B2"),
    ("EfficientNet", "B3"),
    ("EfficientNet", "B4"),
    ("EfficientNet", "B5"),
    ("EfficientNet", "B6"),
    ("EfficientNet", "B7"),
    ("ConvNeXt", "B"),
    ("ConvNeXt", "L")
]

for model, version in model_and_version_list:
    model_data_ord_df.drop(model_data_ord_df[(model_data_ord_df["model"] == model) & (model_data_ord_df["version"] == version)].index, inplace=True)

#

plt.figure(figsize=(20, 10))

for model_idx, model_name in enumerate(model_names):

    color = model_colors[model_name]
    alpha = model_alphas[model_name]

    model_data_filter_df = model_data_ord_df[model_data_ord_df["model"] == model_name]

    plt.scatter(model_data_filter_df["params"], model_data_filter_df["acc@1"], label=model_name, alpha=alpha, color=color)

    for version, params, acc_1 in zip(model_data_filter_df["version"], model_data_filter_df["params"], model_data_filter_df["acc@1"]):
        label = version
        x_pos = params
        y_pos = acc_1 + 0.3
        if model_name == "DeiT" and version in ["B", "B-d", "B-384"]:
            x_pos *= (1.075 + (0.015 * len(version)))
            y_pos -= 0.45
        if model_name == "ViT" and version in ["B/16"]:
            x_pos *= (1.075 + (0.015 * len(version)))
            y_pos -= 0.45
        if model_name == "PVT" and version in ["M", "L"]:
            y_pos -= 0.9
        if model_name == "CvT" and version in ["13-384", "13-384-22k"]:
            x_pos /= (1.075 + (0.015 * len(version)))
            if version == "13-384":
                y_pos -= 0.50
            if version == "13-384-22k":
                y_pos -= 0.40
        if model_name == "SWinT":
            y_pos -= 0.45
            x_pos *= (1.075 + (0.015 * len(version)))
            if version == "B":
                y_pos += 0.20
            if version == "B-V2":
                y_pos += 0.15
        plt.text(x_pos, y_pos, s=label, fontsize=10, ha="center", alpha=alpha)
        
#

plt.xscale("log")

lgnd = plt.legend(loc="lower right")
for handle in lgnd.legend_handles:
    handle._sizes = [30]

plt.xlabel("# Params")
plt.ylabel("ImageNet-1K acc@1")

plt.xlim(model_data_df["params"].min() / 10**(0.1), model_data_df["params"].max() * 10**(0.1))
plt.ylim(model_data_df["acc@1"].min() - 1, model_data_df["acc@1"].max() + 1.5)

#

def circle(xy, radius, kwargs=None):
    """Create circle on figure with axes of different sizes.

    Plots a circle on the current axes using `plt.Circle`, taking into account
    the figure size and the axes units.

    It is done by plotting in the figure coordinate system, taking the aspect
    ratio into account. In this way, the data dimensions do not matter.
    However, if you adjust `xlim` or `ylim` after plotting `circle`, it will
    screw them up; set `plt.axis` before calling `circle`.

    Parameters
    ----------
    xy, radius, kwars :
        As required for `plt.Circle`.

    """

    # Get current figure and axis
    fig = plt.gcf()
    ax = fig.gca()

    # Calculate figure dimension ratio width/height
    pr = fig.get_figwidth()/fig.get_figheight()

    # Get the transScale (important if one of the axis is in log-scale)
    tscale = ax.transScale + (ax.transLimits + ax.transAxes)
    ctscale = tscale.transform_point(xy)
    cfig = fig.transFigure.inverted().transform(ctscale)

    # Create circle
    if kwargs == None:
        circ = mpl.patches.Ellipse(cfig, radius, radius*pr,
                transform=fig.transFigure)
    else:
        circ = mpl.patches.Ellipse(cfig, radius, radius*pr,
                transform=fig.transFigure, **kwargs)

    # Draw circle
    ax.add_artist(circ)

model_and_version_list = [
    ("EfficientNet", "S-V2"),
    ("ViT", "B/16"),
    ("DeiT", "S-d"),
    ("PVT", "S"),
    ("CvT", "21-384-22k"),
    ("SWinT", "T-V2")
]

radius = 0.01

for model, version in model_and_version_list:

    row = model_data_df[(model_data_df["model"] == model) & (model_data_df["version"] == version)]
    params = row.at[row.index[0], "params"]
    acc_1 = row.at[row.index[0], "acc@1"]

    circle((params, acc_1), radius, {"linewidth":2, "edgecolor":model_colors[model], "facecolor":'none', 'clip_on': False})

#

plt.grid()
plt.show()
```
