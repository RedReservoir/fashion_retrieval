```python
def plot_losses(
    exp_name_list,
    exp_data_list
):

    fig, axs = plt.subplots(nrows=2, figsize=(20, 8))

    # ConvNeXt-T Baseline

    exp_data = exp_data_all["ret_train"]["convnext_t"]["test_002"]

    train_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["train_mean_loss_list"]]
    val_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["val_mean_loss_list"]]

    finished = exp_data["results"]["stage_1"]["finished"]

    label_lines = []
    label_lines.append("Stage 1 Baseline")
    label_lines.append("ConvNeXt-T")
    label = "\n".join(label_lines)

    color = [0.66, 0.66, 0.66]

    axs[0].plot(range(1, len(train_loss_list) + 1), train_loss_list, label=label, color=color, marker=".", linestyle="dotted")
    axs[1].plot(range(1, len(val_loss_list) + 1), val_loss_list, label=label, color=color, marker=".", linestyle="dotted")

    # ConvNeXt-T + Fusion Baseline (6)

    exp_data = exp_data_all["ret_train_old_smpl"]["convnext_t_s1"]["exp_006"]

    train_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["train_mean_loss_list"]]
    val_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["val_mean_loss_list"]]

    finished = exp_data["results"]["stage_1"]["finished"]

    label_lines = []
    label_lines.append("Stage 1 Baseline")
    label_lines.append("ConvNeXt-T + Fusion L1234 Head (TopDown)")
    label = "\n".join(label_lines)

    color = [0.33, 0.33, 0.33]

    axs[0].plot(range(1, len(train_loss_list) + 1), train_loss_list, label=label, color=color, marker=".", linestyle="dotted")
    axs[1].plot(range(1, len(val_loss_list) + 1), val_loss_list, label=label, color=color, marker=".", linestyle="dotted")

    # SwinTV2-T Baseline

    exp_data = exp_data_all["ret_train_old_smpl"]["swint_v2_t_s1"]["exp_014"]

    train_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["train_mean_loss_list"]]
    val_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["val_mean_loss_list"]]

    finished = exp_data["results"]["stage_1"]["finished"]

    label_lines = []
    label_lines.append("Stage 1 Baseline")
    label_lines.append("SwinTransformerV2-T")
    label = "\n".join(label_lines)

    color = [0, 0, 0]

    axs[0].plot(range(1, len(train_loss_list) + 1), train_loss_list, label=label, color="black", marker=".", linestyle="dotted")
    axs[1].plot(range(1, len(val_loss_list) + 1), val_loss_list, label=label, color="black", marker=".", linestyle="dotted")

    # ConvNeXt-T + GLAM L4 Baseline

    exp_data_1 = exp_data_all["ret_train_old_smpl_2"]["convnext_t_s1"]["glam_000_exp_006"]

    train_loss_list = [sum(l) for l in exp_data_1["results"]["stage_1"]["train_mean_loss_list"]]
    val_loss_list = [sum(l) for l in exp_data_1["results"]["stage_1"]["val_mean_loss_list"]]

    finished = exp_data["results"]["stage_1"]["finished"]

    label_lines = []
    label_lines.append("Stage 1 Baseline")
    label_lines.append("ConvNeXt-T + GLAM Head")
    label = "\n".join(label_lines)

    color = [0, 0, 0]

    axs[0].plot(range(1, len(train_loss_list) + 1), train_loss_list, label=label, color="black", marker=".", linestyle="dashed")
    axs[1].plot(range(1, len(val_loss_list) + 1), val_loss_list, label=label, color="black", marker=".", linestyle="dashed")

    # Experiments

    for exp_name, exp_data in zip(exp_name_list, exp_data_list):

        # Load data

        train_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["train_mean_loss_list"]]
        val_loss_list = [sum(l) for l in exp_data["results"]["stage_1"]["val_mean_loss_list"]]

        finished = exp_data["results"]["stage_1"]["finished"]
        base_lr = exp_data["settings"]["stage_1"]["optimizer"]["lr"]
        gamma = gamma = exp_data["settings"]["stage_1"]["scheduler"]["gamma"]
        num_gpus = len(exp_data["settings"]["gpu_devices"])
        batch_size = exp_data["settings"]["stage_1"]["data_loading"]["batch_size"]
        grad_acc_iters = exp_data["settings"]["stage_1"]["data_loading"]["grad_acc_iters"]
        backbone_class = exp_data["settings"]["backbone"]["class"]
        head_class = exp_data["settings"]["head"]["class"]

        # Title

        title_lines = []    
        title_lines.append("{:s}".format(exp_name))
        if math.isnan(train_loss_list[-1]):
            title_lines[-1] += " | NAN"
        elif not finished:
            title_lines[-1] += " | Unifinished"
        title_lines.append("{:s} + {:s}".format(backbone_class, head_class))
        title_lines.append("LR: {:s} | GAM: {:.2f} | BS: {:d}".format(
            np.format_float_scientific(base_lr, min_digits=2, exp_digits=1),
            gamma,
            batch_size * num_gpus * grad_acc_iters
        ))
        
        label = "\n".join(title_lines)

        # Early Stopping Mockup

        last_loss = float("inf")
        early_stop_counter = 0
        idx = 0
        while early_stop_counter < 5 and idx < len(val_loss_list):
            if val_loss_list[idx] < last_loss - 1e-10:
                last_loss = val_loss_list[idx]
                early_stop_counter = 0
            else:
                early_stop_counter += 1
            idx += 1

        train_loss_list = train_loss_list[:idx]
        val_loss_list = val_loss_list[:idx]

        # Plot

        axs[0].plot(range(1, len(train_loss_list) + 1), train_loss_list, label=label, marker=".")
        axs[1].plot(range(1, len(val_loss_list) + 1), val_loss_list, label=label, marker=".")

    # Generic settings

    axs[0].set_yscale("log")
    axs[1].set_yscale("log")

    axs[0].grid()
    axs[1].grid()

    axs[0].set_xlabel("Num epochs")
    axs[1].set_xlabel("Num epochs")

    axs[0].set_ylabel("Train loss")
    axs[1].set_ylabel("Val loss")

    plt.legend(prop={'family': 'monospace', "size": 13}, bbox_to_anchor=(1.35, 2.225))
    plt.show()
```

```python
def plot_losses_2(
    exp_name_list_1,
    exp_name_list_2,
    exp_data_list_1,
    exp_data_list_2
):

    fig, axs = plt.subplots(nrows=2, figsize=(20, 12))

    # ConvNeXt-T Baseline S1 + S2

    exp_data = exp_data_all["ret_train"]["convnext_t"]["test_002"]

    train_loss_list_1 = [sum(l) for l in exp_data["results"]["stage_1"]["train_mean_loss_list"]]
    val_loss_list_1 = [sum(l) for l in exp_data["results"]["stage_1"]["val_mean_loss_list"]]

    train_loss_list_2 = [sum(l) for l in exp_data["results"]["stage_2"]["train_mean_loss_list"]]
    val_loss_list_2 = [sum(l) for l in exp_data["results"]["stage_2"]["val_mean_loss_list"]]

    train_loss_list = train_loss_list_1 + train_loss_list_2
    val_loss_list = val_loss_list_1 + val_loss_list_2

    epoch_num_list = [idx + 1 for idx in range(len(train_loss_list_1))] + [idx + 31 for idx in range(len(train_loss_list_2))]

    label_lines = []
    label_lines.append("Stage 1+2 Baseline")
    label_lines.append("ConvNeXt-T")
    label = "\n".join(label_lines)

    color = [0.66, 0.66, 0.66]

    axs[0].plot(epoch_num_list, train_loss_list, label=label, color=color, marker=".", linestyle="dotted")
    axs[1].plot(epoch_num_list, val_loss_list, label=label, color=color, marker=".", linestyle="dotted")

    # ConvNeXt-T + Fusion Baseline (6) S1

    exp_data_1 = exp_data_all["ret_train_old_smpl"]["convnext_t_s1"]["exp_006"]
    exp_data_2 = exp_data_all["ret_train_old_smpl"]["convnext_t_s2"]["exp_006"]

    train_loss_list_1 = [sum(l) for l in exp_data_1["results"]["stage_1"]["train_mean_loss_list"]]
    val_loss_list_1 = [sum(l) for l in exp_data_1["results"]["stage_1"]["val_mean_loss_list"]]

    train_loss_list_2 = [sum(l) for l in exp_data_2["results"]["stage_2"]["train_mean_loss_list"]]
    val_loss_list_2 = [sum(l) for l in exp_data_2["results"]["stage_2"]["val_mean_loss_list"]]

    train_loss_list = train_loss_list_1 + train_loss_list_2
    val_loss_list = val_loss_list_1 + val_loss_list_2

    epoch_num_list = [idx + 1 for idx in range(len(train_loss_list_1))] + [idx + 31 for idx in range(len(train_loss_list_2))]

    label_lines = []
    label_lines.append("Stage 1+2 Baseline")
    label_lines.append("ConvNeXt-T + Fusion Head (TopDown)")
    label = "\n".join(label_lines)

    color = [0.33, 0.33, 0.33]

    axs[0].plot(epoch_num_list, train_loss_list, label=label, color=color, marker=".", linestyle="dotted")
    axs[1].plot(epoch_num_list, val_loss_list, label=label, color=color, marker=".", linestyle="dotted")

    # SwinTV2-T Baseline S1 + S2

    exp_data_1 = exp_data_all["ret_train_old_smpl"]["swint_v2_t_s1"]["exp_014"]
    exp_data_2 = exp_data_all["ret_train_old_smpl"]["swint_v2_t_s2"]["run_002"]

    train_loss_list_1 = [sum(l) for l in exp_data_1["results"]["stage_1"]["train_mean_loss_list"]]
    val_loss_list_1 = [sum(l) for l in exp_data_1["results"]["stage_1"]["val_mean_loss_list"]]

    train_loss_list_2 = [sum(l) for l in exp_data_2["results"]["stage_2"]["train_mean_loss_list"]]
    val_loss_list_2 = [sum(l) for l in exp_data_2["results"]["stage_2"]["val_mean_loss_list"]]

    train_loss_list = train_loss_list_1 + train_loss_list_2
    val_loss_list = val_loss_list_1 + val_loss_list_2

    epoch_num_list = [idx + 1 for idx in range(len(train_loss_list_1))] + [idx + 31 for idx in range(len(train_loss_list_2))]

    label_lines = []
    label_lines.append("Stage 1+2 Baseline")
    label_lines.append("SwinTransformerV2-T")
    label = "\n".join(label_lines)

    color = [0, 0, 0]

    axs[0].plot(epoch_num_list, train_loss_list, label=label, color=color, marker=".", linestyle="dotted")
    axs[1].plot(epoch_num_list, val_loss_list, label=label, color=color, marker=".", linestyle="dotted")

    # ConvNeXt-T + GLAM L4 Baseline

    exp_data_1 = exp_data_all["ret_train_old_smpl_2"]["convnext_t_s1"]["glam_000_exp_006"]
    exp_data_2 = exp_data_all["ret_train_old_smpl_2"]["convnext_t_s2"]["glam_000_exp_007"]

    train_loss_list_1 = [sum(l) for l in exp_data_1["results"]["stage_1"]["train_mean_loss_list"]]
    val_loss_list_1 = [sum(l) for l in exp_data_1["results"]["stage_1"]["val_mean_loss_list"]]

    train_loss_list_2 = [sum(l) for l in exp_data_2["results"]["stage_2"]["train_mean_loss_list"]]
    val_loss_list_2 = [sum(l) for l in exp_data_2["results"]["stage_2"]["val_mean_loss_list"]]

    train_loss_list = train_loss_list_1 + train_loss_list_2
    val_loss_list = val_loss_list_1 + val_loss_list_2

    epoch_num_list = [idx + 1 for idx in range(len(train_loss_list_1))] + [idx + 31 for idx in range(len(train_loss_list_2))]

    label_lines = []
    label_lines.append("Stage 1 Baseline")
    label_lines.append("ConvNeXt-T + GLAM Head")
    label = "\n".join(label_lines)

    color = [0, 0, 0]

    axs[0].plot(epoch_num_list, train_loss_list, label=label, color=color, marker=".", linestyle="dashed")
    axs[1].plot(epoch_num_list, val_loss_list, label=label, color=color, marker=".", linestyle="dashed")

    # Experiments

    for exp_name_1, exp_data_1, exp_name_2, exp_data_2 in zip(exp_name_list_1, exp_data_list_1, exp_name_list_2, exp_data_list_2):

        # Load data

        if exp_data_1 is None:
            train_loss_list_1 = []
            val_loss_list_1 = []
        else:
            train_loss_list_1 = [sum(l) for l in exp_data_1["results"]["stage_1"]["train_mean_loss_list"]]
            val_loss_list_1 = [sum(l) for l in exp_data_1["results"]["stage_1"]["val_mean_loss_list"]]
            
        train_loss_list_2 = [sum(l) for l in exp_data_2["results"]["stage_2"]["train_mean_loss_list"]]
        val_loss_list_2 = [sum(l) for l in exp_data_2["results"]["stage_2"]["val_mean_loss_list"]]

        train_loss_list = train_loss_list_1 + train_loss_list_2
        val_loss_list = val_loss_list_1 + val_loss_list_2

        epoch_num_list = [idx + 1 for idx in range(len(train_loss_list))]

        finished = exp_data_2["results"]["stage_2"]["finished"]
        base_lr = exp_data_2["settings"]["stage_2"]["optimizer"]["lr"]
        gamma = gamma = exp_data_2["settings"]["stage_2"]["scheduler"]["gamma"]
        num_gpus = len(exp_data_2["settings"]["gpu_devices"])
        batch_size = exp_data_2["settings"]["stage_2"]["data_loading"]["batch_size"]
        grad_acc_iters = exp_data_2["settings"]["stage_2"]["data_loading"]["grad_acc_iters"]
        backbone_class = exp_data_2["settings"]["backbone"]["class"]
        head_class = exp_data_2["settings"]["head"]["class"]

        # Title

        title_lines = []    
        title_lines.append("{:s}".format(exp_name_2))
        if math.isnan(train_loss_list[-1]):
            title_lines[-1] += " | NAN"
        elif not finished:
            title_lines[-1] += " | Unifinished"
        title_lines.append("{:s} + {:s}".format(backbone_class, head_class))
        title_lines.append("LR: {:s} | GAM: {:.2f} | BS: {:d}".format(
            np.format_float_scientific(base_lr, min_digits=2, exp_digits=1),
            gamma,
            batch_size * num_gpus * grad_acc_iters
        ))
        label = "\n".join(title_lines)

        
        # Plot

        axs[0].plot(epoch_num_list, train_loss_list, label=label, marker=".")
        axs[1].plot(epoch_num_list, val_loss_list, label=label, marker=".")

    # Generic settings

    axs[0].set_yscale("log")
    axs[1].set_yscale("log")

    axs[0].grid()
    axs[1].grid()

    axs[0].set_xlabel("Num epochs")
    axs[1].set_xlabel("Num epochs")

    axs[0].set_ylabel("Train loss")
    axs[1].set_ylabel("Val loss")

    plt.legend(prop={'family': 'monospace', "size": 13}, bbox_to_anchor=(1.30, 2.225))
    plt.show()
```
